# 2020년 건강검진 일부 데이터 -> 산점도 그래프

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib
```

```python
# 데이터 불러오기
data = pd.read_excel('health_screenings_2020_1000ea.xlsx')
data.head()
```

## 키, 몸무게 데이터 -> 산점도 그래프

```python
heightdata = data['height']
weightdata = data['weight']

plt.figure(figsize=(10, 4))

plt.scatter(heightdata, weightdata)

plt.title('2020년 건강검진 산점도 그래프')
plt.xlabel('키')
plt.ylabel('몸무게')
plt.grid()

plt.show()
```

## 수축기 혈압, 이완기 혈압, 공복 혈당 그룹 산점도 그래프

```python
systolic_data = data['systolic']
diastolic_data = data['diastolic']
blood_sugar_data = data['blood_sugar']

plt.figure(figsize=(10, 6))

plt.scatter(systolic_data, diastolic_data, color='r', edgecolors='w', label='수축기*이완기')
plt.scatter(systolic_data, blood_sugar_data, color='g', edgecolors='w', label='수축기*공복혈당')
plt.scatter(diastolic_data, blood_sugar_data, color='b', edgecolors='w', label='이완기*공복혈당')

plt.grid()
plt.title('2020년 건강검진 그룹 산점도 그래프')
plt.legend()
plt.xlim(40, 180)
plt.ylim(30, 300)

plt.show()
```

## 여성 키 히스토그램

```python
data.head()
```

```python
female_data = data.loc[data['gender']==2, ['gender', 'height']]
female_data['height']
```

```python
plt.figure(figsize=(10, 6))

plt.hist(female_data['height'], bins=7, label='female') # bins : 구간 단위

plt.title('2020년 건강검진 여성 키 히스토그램')
plt.xlabel('키')
plt.ylabel('빈도수(frequency)')
plt.xlim(130, 180)
plt.grid()
plt.legend()

plt.show()
```

## 남성, 여성 키 그룹 히스토그램

```python
data.head()
```

```python
male_data = data.loc[data['gender']==1, ['gender', 'height']]
male_data.head()
```

```python
plt.figure(figsize=(10, 6))

plt.hist(male_data['height'], bins=7, alpha=0.5, label='남성') # alpha : 투명도
plt.hist(female_data['height'], bins=7, alpha=0.5, label='여성') # alpha=0.5 : 투명도 50%
         
plt.title('2020년 건강검진 남성 키, 여성 키 그룹 히스토그램')
plt.xlabel('키')
plt.ylabel('빈도수')
plt.legend()
plt.grid()
plt.xlim(130, 195)

plt.show()
```

## 남성, 여성 키 상자수염 그래프

```python
import numpy as np

# 배열
height = np.array([male_data['height'], female_data['height']], dtype=object) # dtype=object : 데이터 타입이 문자열
height
```

```python
import matplotlib
matplotlib.__version__
```

```python
plt.figure(figsize=(8, 5))

plt.boxplot(height)

plt.ylim(130, 200)
plt.xlabel('gender')
plt.ylabel('height')
plt.title('2020년 건강검진 남성 키, 여성 키 상자수염 그래프')
plt.xticks([1, 2], ['남', '여'])

plt.show()
```
