```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import koreanize_matplotlib
import seaborn as sns
```

# 4번

```python
data = pd.read_excel('health_screenings_2020_1000ea.xlsx')\

data.head()
```

```python
data4 = data.loc[:, ['city_code', 'gender', 'waist']]

male_data = data4[data4['gender']==1].groupby('city_code')['waist'].mean()
female_data = data4[data4['gender']==2].groupby('city_code')['waist'].mean()
```

```python
plt.figure(figsize=(12, 4))

plt.plot(male_data, linestyle='--', color='b', label='Male', marker='o')
plt.plot(female_data, linestyle='--', color='r', label='Female', marker='o')

plt.xlabel('Count')
plt.ylabel('Waist')
plt.title('2020 Health Screenings Waist')
plt.legend()
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10, 5))

plt.bar(waistData.index, waistData['waist'], color='c')

plt.ylim(50, 110)
plt.xlabel('Count')
plt.ylabel('Waist')
plt.title('2020 Health Screenings Waist')
plt.grid()

plt.show()
```

# 7번

```python
maledata = data.loc[data['gender']==1, ['gender', 'waist']]
femaledata = data.loc[data['gender']==2, ['gender', 'waist']]
```

```python
plt.figure(figsize=(10, 6))

plt.hist(maledata['waist'], bins=8, label='Male', alpha=0.5)
plt.hist(femaledata['waist'], bins=8, label='Female', alpha=0.5)

plt.xlim(50, 115)
plt.ylim(0, 200)
plt.xlabel('Waist')
plt.ylabel('Frequency')
plt.legend()
plt.grid()

plt.show()
```
