```python
import os
from dotenv import load_dotenv

load_dotenv()

API_KEY = os.getenv('OPENAPI_API_KEY')
```

```python
# 공공데이터 포털의 api 인증키 인식이 잘 되는지 테스트

import requests # 웹페이지의 내용(html)을 쉽게 가지고 오도록 해주는 라이브러리
import json

# .strip() : 양쪽 공백 제거

print(f'API key 길이 : {len(API_KEY)}')
print(f'앞 10자리 : {API_KEY[:10]}')
```

```python
url = 'https://apis.data.go.kr/6270000/dbmsapi02/getBasic02'

params = {'serviceKey':API_KEY}

response = requests.get(url, params=params)

print(f'응답코드 : {response.status_code}')
```

```python
# json으로 변환
data = response.json()

data
```

```python
data.keys()
```

```python
list(data.keys())
```

```python
# 헤더 (성공 여부)
data['header']
```

```python
print(f'성공 여부 : {data["header"]["success"]}')
```

```python
print(f"결과 메시지 : {data['header']['resultMsg']}")
```

```python
# 실제 데이터 (body)
body = data['body']
body
```

```python
print(f"총 데이터 수 : {body['totalCount']}")
```

```python
print(f'items 키 목록 : {list(body["items"].keys())}')
```

```python
# 노선 데이터
routes = body['items']['route']
routes
```

```python
print(f'전체 노선 수 : {len(routes)}')
```

```python
print(f'첫 번째 노선 정보 : {routes[0]}')
```

```python
# 급행 노선만 찾기

for route in routes:
    if '급행' in route['routeNo']:
        print(f'{route["routeNo"]}\t번 버스')
```

```python
for route in routes:
    if '급행' in route['routeNo']:
        print(f'{route["routeNo"]}\t번 버스 : {route["stNm"]} -> {route["edNm"]}')
```

```python
# 버스 정류장
stops = body['items']['bs']
stops[0]
```

```python
# 동대구역이 포함된 정류장 모두 출력

keyword = '동대구역'

for stop in stops:
    if keyword in stop['bsNm']:
        print(f"정류장 이름 : {stop['bsNm']}")
```
