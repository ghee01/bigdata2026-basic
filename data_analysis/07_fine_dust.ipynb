# 에어코리아 - 미세먼지 (2001년 ~ 2019년)

## fine_dust.xlsx 파일 불러오기, 라이브러리 불러오기

```python
import pandas as pd
import matplotlib.pyplot as plt
import koreanize_matplotlib

data = pd.read_excel('fine_dust.xlsx')
data
```

## 2018년 지역별 미세먼지 선형 그래프

```python
data2018 = data[2018]
data2018
```

```python
plt.figure(figsize=(15, 4)) # 15X4(inch) 사이즈의 도화지 생성
plt.plot(data2018, color='m', marker='o')
plt.title('2018년 미세먼지 선 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.grid() # 격자무늬
plt.show()
```

## 2016년 ~ 2019년 지역별 미세먼지 선형 그래프

```python
for year in range(2016,2020) :
    chartdata = data[year]
    print(chartdata)
    print('-' * 30)
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016,2020) :
    chartdata = data[year]
    plt.plot(chartdata, marker='s', label=year)

plt.title('2016 ~ 2019 미세먼지 선형 그래프')
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.legend()
plt.grid()
plt.show()
```

## 2017년 지역별 미세먼지 세로 막대 그래프

- plt.bar(x=x data, height=y data)

```python
plt.figure(figsize=(15,4))
plt.bar(data['area'], data[2017], color='skyblue')
plt.title('2017년 미세먼지 막대 그래프')
plt.ylim(35, 55)
plt.show()
```

## 2016년 ~ 2019년 경기, 인천, 부산, 대구 지역별 미세먼지 그룹 세로 막대 그래프

```python
data
```

```python
data = pd.read_excel('fine_dust.xlsx', index_col=0)
data
```

```python
data4 = data.loc['Gyeonggi':'Daegu', 2016:]
data4
```

```python
plt.figure(figsize=(15, 4))
for year in range(2016, 2020) :
    chartdata = data4[year]
    plt.bar(data4.index, chartdata, label=year)

plt.show()
```

```python
import numpy as np

index = np.arange(4) # 0~3

plt.figure(figsize=(15, 4))
for year in range(2016, 2020) :
    chartdata = data4[year]
    plt.bar(index, chartdata, width=0.2, label=year)
    index = index + 0.2

plt.show()
```

```python
import numpy as np

index = np.arange(4) # 0~3

plt.figure(figsize=(15, 4))
for year in range(2016, 2020) :
    chartdata = data4[year]
    plt.bar(index, chartdata, width=0.2, label=year)
    index = index + 0.2

plt.title('2016~2019 그룹별 미세먼지 세로 막대 그래프')
plt.ylim(35,55)
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.legend()

plt.show()
```

```python
import numpy as np

index = np.arange(4) # 0~3

plt.figure(figsize=(15, 4))
for year in range(2016, 2020) :
    chartdata = data4[year]
    plt.bar(index, chartdata, width=0.2, label=year)
    index = index + 0.2

plt.title('2016~2019 그룹별 미세먼지 세로 막대 그래프')
plt.ylim(35,55)
plt.xlabel('지역')
plt.ylabel('micrometer')
plt.xticks(index-0.5, ['경기', '인천', '부산', '대구']) # x축 글자 다시 입력
plt.legend()

plt.show()
```
