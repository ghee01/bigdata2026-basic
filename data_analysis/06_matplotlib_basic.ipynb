# Matplotlib 라이브러리 기본 시각화

```python
# 라이브러리 불러오기
import matplotlib.pyplot as plt
```

```python
# 데이터
xdata = [2, 3, 4, 5]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata) # 선형그래프
plt.title('Line Graph')
plt.show()
```

## 범례 -> plt.legend()

```python
data1 = [2, 3, 4, 5]
data2 = [8, 6, 4, 2]
plt.plot(data1, label='asc')
plt.plot(data2, label='desc')
plt.legend() # 범례 추가
plt.show()
```

## 선 색상 지정

```python
plt.plot(xdata, ydata, color='red')
plt.show()
```

## 축 제목

-plt.xlabel()
-plt.ylabel()

```python
plt.plot(xdata, ydata, color='g')
plt.xlabel('x축 제목')
plt.ylabel('y축 제목')
plt.show
```

## 선 종류 지정

- linestyle='dashed' | linestyle='--'

```python
plt.plot(data1, color='r', label='dashed', linestyle='--') # 파선
plt.plot(data2, color='g', label='dotted', linestyle=':') # 점선
plt.plot(ydata, color='b', label='dash-dot', linestyle='-.') # 1점쇄선
plt.legend()
plt.show()
```

## 마커

```python
plt.title('X, Y range')
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='c', linestyle='--', marker='o', markersize='10')
plt.show()
```

## 축 범위 조정

- plt.xlim(범위)
- plt.xylim(범위)

```python
plt.title('X, Y range')
xdata = [5, 10, 15, 20]
ydata = [1, 3, 5, 7]
plt.plot(xdata, ydata, color='c', linestyle='--', marker='o', markersize='10')
plt.xlim(0, 25)
plt.ylim(-2, 10)
plt.show()
```
