### 결측치 처리

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 결측 비율이 50% 초과 시, 컬럼 삭제
# 평균 : 결측 비율을 제대로 알기 위해 평균을 구한다
missing_ratio = df.isnull().mean()
missing_ratio
```

```python
missing_ratio[missing_ratio > 0.5]
```

```python
drop_cols = missing_ratio[missing_ratio > 0.5].index.tolist()
```

```python
df.drop(columns=drop_cols)
```
