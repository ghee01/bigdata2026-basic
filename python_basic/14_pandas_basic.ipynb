## bool 인덱싱

```python
w = df['temp'] >= 30
df[w] # True만 추출
```

```python
df[df['temp'] >= 30]
```

### 조건에 맞는 데이터 추출

```python
# 온도가 가장 높은 날의 모든 정보(행)

w = df['temp'] == df['temp'].max()
df[w]
```

### 조건 2개 이상

- and - &
- or - |

```python
# 기온이 30도 이상이고, 최대 풍속이 9 이상인 데이터

w = (df['temp'] >= 30) | (df['max_wind'] >= 9)
df[w]
```

# 판다스 결측데이터 처리

## isna() : 값이 nan(결측치) -> True

```python
# 최대 풍속이 측정되지 않은 모든 데이터 출력

df[df['max_wind'].isna()]
```

## 데이터프레임명.isnull().sum() : 컬럼별 결측데이터 개수

- isnull의 값
    - 0 : False
    - 1 : True

```python
df.isnull().sum() # 컬럼별
```

```python
df.isnull().sum().sum() # 전체 데이터프레임
```

## dropna() : 결측데이터 삭제

```python
df.dropna() # 결측치 제거한 상태 확인
```

```python
df # 원본의 결측치가 제거되지 않음
```

```python
df2 = df.dropna() # 결측치 제거한 상태를 새로운 데이터프레임에 저장
df2
```

```python
df2.isnull().sum()
```

```python
df.isnull().sum()
```

## fillna(대체값) : 결측데이터 대체

```python
df['max_wind'].isnull().sum()
```

```python
df3 = df.copy() # 복사
df3
```

```python
df3 = df3.fillna(0) # 결측치 모두 0으로 채움
```

```python
df3.isnull().sum()
```

```python
# 평균

avg = df3['max_wind'].mean()
avg
```

```python
df3['max_wind'] = df3['max_wind'].fillna(avg)
```

```python
avg2 = df3['mean_wind'].mean()
avg2
```

```python
df3['mean_wind'] = df3['mean_wind'].fillna(avg2)
```
