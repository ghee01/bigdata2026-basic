### 기온(temp) 컬럼의 내용

- 데이터프레임명[컬럼명].메서드명()

```python
df['temp']
```

```python
print(df['temp'].mean()) # 평균
```

```python
df['temp'].sum() # 합계
```

```python
df['temp'].median() # 중간값
```

## sort_values() : 데이터 크기순 정렬

- ascending = True (생략) -> 오름차순
- ascending = False -> 내림차순
- by = [컬럼명] : 기준

```python
# 최대 풍속(max_wind) 데이터 크기에 따라 오름차순 정렬

df.sort_values(by='max_wind')
```

```python
# 최대 풍속(max_wind) 데이터 크기에 따라 내림차순 정렬

df.sort_values(by='max_wind', ascending=False)
```

## value_counts() : 빈도 분석 결과

```python
filename = 'bank.csv'
bank = pd.read_csv(filename)
bank.head()
```

```python
bank['job'].value_counts() # 기본값 : ascending=False
```

```python
bank['job'].value_counts(ascending=True)
```
