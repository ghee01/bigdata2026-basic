# 대구광역시 달서구 cctv 개수 현황

```python
import csv

with open('cctv_20250527.csv', 'r') as f :
    buffer = csv.reader(f, delimiter=',')
    totalcctv = 0 # cctv 개수

    for i, line in enumerate(buffer) : # i : 인덱스 번호 / line : 값
        if i != 0 :
            totalcctv += int(line[6])
```

# 부산 cctv 현황

```python
import csv

with open('cctv_busan.csv', 'r') as f :
    buffer = csv.reader(f, delimiter=',')
    
    line1 = 0
    line2 = 0
    line3 = 0
    line4 = 0

    for i, line in enumerate(buffer) :
        if i != 0 :
            if int(line[0]) == 1 :
                line1 += sum(int(x) for x in line[2:6])
            elif int(line[0]) == 2 :
                line2 += sum(int(x) for x in line[2:6])
            elif int(line[0]) == 3 :
                line3 += sum(int(x) for x in line[2:6])
            else :
                line4 += sum(int(x) for x in line[2:6])
```

```python
print('부산 지하철 호선별 CCTV 개수')
print('-' * 30)
print(f'1호선: {line1}개')
print(f'2호선: {line2}개')
print(f'3호선: {line3}개')
print(f'4호선: {line4}개')
```
