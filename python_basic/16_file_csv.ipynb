# csv 파일 읽기

```python
result_list = []

with open('../data_analysis/bank.csv', 'rt') as f :
    while True :
        line = f.readline() # job,education,balance,duration,y
        
        if not line :
            break

        result = line.split(',') # ['job','education','balance','duration','y']
        result_list.append(result) # [['job','education','balance','duration','y']]

result_list
```

```python
result_list = []

with open('../data_analysis/bank.csv', 'rt') as f :
    f.readline() # job,education,balance,duration,y # 제목 줄 읽고 날림
    while True :
        line = f.readline() # management,tertiary,2143,261,no
        
        if not line :
            break

        result = line.split(',') # ['management','tertiary','2143','261','no']
        result_list.append(result) # [['management','tertiary','2143','261','no']]

result_list
```
