# 판다스 데이터 가공

```python
import pandas as pd
import numpy as np

filename = 'dust1.xlsx'
dust = pd.read_excel(filename)
dust.head()
```

## 컬럼(열, 변수, 피처) 삭제 / 생성

### drop(컬럼명, axis=1) : 열 삭제

- drop(..., [axis=0]) : 행 삭제

```python
# 지역, 망, 측정소코드 삭제

dust.drop(['지역', '망', '측정소코드'], axis=1) # 확인 용도
```

```python
dust = dust.drop(['지역', '망', '측정소코드'], axis=1) # 덮어쓰기
dust.head()
```

### 컬럼 생성(추가)

- 데이터프레임명[컬럼명] = 값

```python
dust['city'] = '서울'
```

```python
dust.head()
```

## rename() : 컬럼명 수정

- 데이터프레임명.rename(columns={'기존 이름':'새 이름'})

```python
dust = dust.rename(columns={
    '측정소명':'name',
    '측정일시':'date',
    '주소':'addr'
})
```

```python
dust.columns
```

## dtypes : 데이터 타입 확인

```python
dust.dtypes # 판다스 버전 3부터 object -> str 변경
```

## astype() : 데이터 형변환

- [컬럼명].astype(변환할 자료형)

```python
dust['date'].astype(str) # 확인
```

```python
dust['date'] = dust['date'].astype(str) # 덮어쓰기
```

```python
dust.dtypes
```

```python
dust['date']
```

```python
# 첫 8자리만 자른다 -> slice()

dust['date'] = dust['date'].str.slice(0, 8)
```

## to_datetime() : 날짜/시간 형변환 (시계열)

```python
pd.to_datetime(dust['date'])
```

```python
dust['date'] = pd.to_datetime(dust['date'])
dust.dtypes
```
