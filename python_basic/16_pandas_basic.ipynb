```python
# 라이브러리 불러오기

import pandas as pd
import numpy as np
```

```python
dust = pd.read_excel('dust1.xlsx')
dust.head()
```

```python
# 컬럼 이름 변경

dust = dust.rename(columns={'측정일시':'date'})
dust.head()
```

```python
# 자료형 확인

dust.dtypes
```

```python
# 문자열로 형변환

dust['date'] = dust['date'].astype(str)
dust['date'] = dust['date'].str.slice(0,8)
dust.dtypes
```

```python
# 날짜형으로 형변환

dust['date'] = pd.to_datetime(dust['date'])
dust.dtypes
```

# 날짜형을 년, 월, 일로 분리 추출

```python
# dt 속성

dust['date'].dt
```

```python
dust['date'].dt.year # 연도 추출
```

## 년, 월, 일 컬럼 생성

```python
dust['year'] = dust['date'].dt.year
dust['month'] = dust['date'].dt.month
dust['day'] = dust['date'].dt.day
```

```python
dust.head()
```

# 데이터 병합

```python
s1 = pd.read_excel('nation.xlsx')
s2 = pd.read_excel('code.xlsx')
```

```python
s1.head()
```

```python
s2.head()
```

## merge()

- pd.merge(데이터프레임1, 데이터프레임2, on=기준컬럼명, how='조인형식')
- how : inner(기본값), outer, left, right

```python
# 국적 코드 컬럼 기준

pd.merge(s1, s2, on='국적코드')
```

```python
# 오른쪽 컬럼 기준(차집합)

pd.merge(s1, s2, on='국적코드', how='right')
```

# 데이터 그룹핑

## groupby()

```python
# 국적코드별 성별, 입국객수, 증가수

s1.groupby('국적코드').sum()
```

```python
# 국적코드, 성별 -> 입국객수, 증가수

s1.groupby(['국적코드', '성별']).sum()
```
