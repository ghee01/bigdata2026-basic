## 2.

```python
guest = []

with open ('guest.txt', 'w') as f :
    while True :
        name = input('name : ')

        if not name :
            break

        f.write(name + '\n')
        guest.append(name)

        print(f'{name} welcome!')

print(f'total count : {len(guest)}')
```

## 3.

```python
with open('data2.txt', 'w') as f :
    f.write('98\n')
    f.write('86\n')
    f.write('75\n')
    f.write('47\n')
    f.write('83')
```

```python
score = []

with open('data2.txt', 'r') as f :
    while True :
        line = f.readline()

        if not line :
            break

        score.append(int(line.rstrip('\n')))
score
```

```python
sum = sum(score)
avg = sum / len(score)

print(f'점수 합계 : {sum}')
print(f'점수 평균 : {avg}')
```

```python
import json

mt = {'가평':0,
      '대성리':0,
      '남이섬':0,
      '청평':0}
print(mt)

# 파일 쓰기
with open('mt.json', 'w', encoding='utf-8') as f :
    print('Best MT Spot')
    while True :
        vote = input("area : ")
        if not vote :
            break
        mt[vote] += 1

    f.write(json.dumps(mt))

# 파일 읽기
with open('mt.json', 'r', encoding='utf-8') as f :
    res_json = f.read()
    res_dic = json.loads(res_json)

res_dic
```
