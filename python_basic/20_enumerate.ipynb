# 리스트의 for문

```python
ninja = ['나루토', '시카마루', '키바', '쵸지', '네지']
```

```python
for n in ninja :
    print(n) # 리스트의 값(요소) 출력
```

```python
for i in range(5) :
    print(i)
```

# enumerate 함수

for 인덱스번호, 값 in enumerate(컬렉션자료형) :

```python
for n in enumerate(ninja) :
    print(n)
```

## 튜플 언패킹

각각의 변수에 튜플의 값을 나눠 담는다

```python
for index, name in enumerate(ninja) :
    print(f'{index+1}번: {name}')
```

```python
for index, name in enumerate(ninja, start=1) :
    print(f'{index}번: {name}')
```
