# JSON
- 자바스크립트로 만들어진 경량의 데이터 파일
- {"key":"value"} --> 파이썬의 딕셔너리와 비슷하다
- 변환이 쉽다

```python
import json
```

- 파이썬으로 딕셔너리 --> json 변환

```python
dict_list = [
    {
        'name':'james',
        'age':20,
        'spec':[175.5, 70.5]
    },
    {        
        'name':'alice',
        'age':21,
        'spec':[168.5, 60.5]
    }
]
```

```python
dict_list
```

```python
type(dict_list)
```

```python
type(dict_list[0])
```

## JSON으로 변환

```python
json_string = json.dumps(dict_list)
```

```python
json_string
```

```python
type(json_string)
```

## JSON 파일로 저장 (JSON으로 내보내기)

```python
with open('dataFile/dictList.json', 'w') as f :
    f.write(json_string)
```

## JSON 파일 읽기

```python
import json

with open('dataFile/dictList.json', 'r') as f :
    json_reader = f.read()
    dict_list = json.loads(json_reader) # 파이썬으로 변환
```

```python
dict_list
```

```python
type(dict_list)
```

```python
for dic in dict_list :
    print(f'이름 : {dic["name"]}')
    print(f'나이 : {dic["age"]}')
    print(f'키 : {dic["spec"][0]}')
    print(f'몸무게 : {dic["spec"][1]}')
    print('-' * 20)
```

## 실습 - 서울시 마포구의 cctv 설치 목적

```python
import json

with open('dataFile/cctv.json', 'r', encoding='utf-8') as f :
    buffer = f.read()
    cctv_list = json.loads(buffer) # 파이썬으로 변환
    cctv_purpose = set() # 중복 제거 위한 빈 세트 생성

    for place in cctv_list :
        cctv_purpose.add(place['설치목적구분']) # add : set에 데이터 추가하는 메서드
```
