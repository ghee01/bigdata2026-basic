# 클래스와 클래스 변수

## 클래스 변수

- 여러 인스턴스가 변수 공간을 같이 사용하는 변수
- 같은 클래스를 통해서 만들어진 인스턴스들끼리 공간을 사용

```python
class Person :
    population = 0 # 클래스 변수 -> 인구 수 저장

    def __init__(self, name):
        """
        생성자 : 인스턴스가 생성되면 인구수 1명 증가
        """
        self.name = name # 인스턴스 변수(self.name)
        Person.population += 1 # 인구수 1 증가

    def print_hobby(self, hobby) :
        """취미를 알려주는 인스턴스 메서드"""
        print(f'{self.name}이(가) {hobby}을(를) 즐깁니다')
```

```python
saram1 = Person('청명') # 인스턴스 생성
saram2 = Person('백천')
saram3 = Person('유이설')
saram4 = Person('윤종')
saram5 = Person('조걸')
```

```python
saram1.print_hobby('음주')
saram2.print_hobby('비무')
saram3.print_hobby('수련')
saram4.print_hobby('명상')
saram5.print_hobby('독서')
```

```python
print(f'생성된 사람들은 모두 {Person.population}명입니다')
```

```python
saram6 = Person('당소소')
saram6.print_hobby('유이설 좋아하기')
```

```python
print(f'생성된 사람들은 모두 {Person.population}명입니다')
```

## 클래스 메서드 (class method)

- 인스턴스(객체)가 아니라 클래스 자체를 대상으로 동작하는 메서드
- 클래스 변수 수정
- 클래스 전체 설정 변경
- @classmethod 데코레이터 사용
- 첫번째 매개변수로 cls 사용
- 클래스 메서드는 인스턴스를 만들지 않고 바로 클래스명.클래스메서드()로 호출 가능

### 사용하는 이유

- 보통 여러 인스턴스가 있을 때 공통 데이터를 관리하고자 할 때 사용
ex) 학교 이름, 회사 이름, 환율 설정, 게임 서버 상태, 환경 설정 값 등

```python
class Student:
    school = '유에이 고교' # 클래스 변수

    def instance_method(self) :
        print('self가 있으니 인스턴스 메서드')

    @classmethod
    def class_method(cls) :
        print('cls가 있으니 클래스 메서드')
```

```python
Student.class_method()
```

```python
s1 = Student()
s1.instance_method()
```

```python
s1.class_method() # 가능
```

```python
# 클래스 메서드를 추가한 수정ver.

class Person :
    population = 0 # 클래스 변수 -> 인구 수 저장

    def __init__(self, name):
        """
        생성자 : 인스턴스가 생성되면 인구수 1명 증가
        """
        self.name = name # 인스턴스 변수(self.name)
        Person.population += 1 # 인구수 1 증가

    def print_hobby(self, hobby) :
        """취미를 알려주는 인스턴스 메서드"""
        print(f'{self.name}이(가) {hobby}을(를) 즐깁니다')

    @classmethod
    def get_population(cls) :
        """인구수를 반환해주는 클래스 메서드"""
        return cls.population
```

```python
print(f'현재 인구 수는 {Person.get_population()}명입니다')
```

```python
saram1 = Person('청명')
```

```python
print(f'현재 인구 수는 {Person.get_population()}명입니다')
```

```python
saram2 = Person('백천')
saram3 = Person('유이설')
```

```python
print(f'현재 인구 수는 {Person.get_population()}명입니다')
```

### 실습 예제 - 분식집 -> 메뉴, 개수 -> 총 금액

```python
class Korean_snack :
    total = 0

    menu_list = [
        # 딕셔너리 {'메뉴' : 가격}
        {'떡볶이':3000},
        {'순대':3000},
        {'튀김':500},
        {'김밥':2000}
    ] # 메뉴판 (클래스 변수)

    @classmethod
    def sales(cls, food, count) :
        """
        총 계산 클래스 메서드
        
        매개변수(파라미터)
        food : 음식 이름
        count : 개수
        """
        for menu in cls.menu_list : # menu = {'떡볶이':3000}
            if food in menu :
                cls.total += count * menu[food]
```

```python
Korean_snack.sales('떡볶이', 2)
Korean_snack.sales('순대', 1)
Korean_snack.sales('튀김', 5)
```

```python
print(f'분식집에 내야 할 금액 : {Korean_snack.total}원')
```
